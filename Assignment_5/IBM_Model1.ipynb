## IBM Model 1

Target Language: French  
French Sentence:   
**f** <f1, ..., f_i> with size **I**   


Source Language: English  
English Sentence:  
**e** <e1, ..., e_j> with size **J**  
  
Alingment of sentences:  
Array **a**  <a1, ..., a_i> of length **I** 
The values of **a** correspond to the index of the english words, it's index is the index of the french word.  


```python
import pandas as pd
import numpy as np
```

```python
# Number of sentences
n_sents = 10
```

```python
# Loading french sentences
with open ('./hw2/data/hansards.f', 'r') as file:
    f_sentences = file.readlines()[:n_sents]
print(f_sentences[1])
```

```python
# Loading english sentences
with open ('./hw2/data/hansards.e', 'r') as file:
    e_sentences = file.readlines()[:n_sents]
print(e_sentences[1])
```

```python
# Selecting only the first 1000 sentences of the Hansard corpus
vocab_french = set()
vocab_english = set()
for i in range(n_sents):
    vocab_french.update(f_sentences[i].split())
    vocab_english.update(e_sentences[i].split())
```

```python
# Initializing the a table of counts p(f,e)

d = np.ones(shape=(len(vocab_french),len(vocab_english)))

table = pd.DataFrame(d, columns=vocab_english, index=vocab_french)

for i in range(len(vocab_french)):
    for j in range(len(vocab_english)):
        table.iloc[i][j] = table.iloc[i][j]/len(vocab_english)
```

```python
for step in range(5):
    d = np.zeros(shape=(len(vocab_french),len(vocab_english)))
    counts = pd.DataFrame(d, columns=vocab_english)
    total = np.zeros(len(vocab_french))
    Z = {}
    
    for n in range(n_sents):
        e = e_sentences[n].split()
        for j in e:
            # ei = index o the english word in the table table
            ei = table.columns.get_loc(j)
            Z[j] = 0
            
            f = f_sentences[n].split()
            for k in f:
                # fi = index of the french word in the table
                fi = np.where(table.index==k)[0][0]              
                Z[j] = Z[j] + table.iloc[fi][ei]

        for j in e:
            ei = table.columns.get_loc(j)
                    
            for k in f:
                fi = np.where(table.index==k)[0][0]
                
                # Expected count
                c = table.iloc[fi,ei]/Z[j]
                
                counts.iloc[fi,ei] += c
                total[fi] += c

    for i in range(len(vocab_french)):
        for j in range(len(vocab_english)):
            table.iloc[i][j] = counts.iloc[i][j]/total[i]
            
```
